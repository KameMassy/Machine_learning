# What is Batch(offline)  learning?

###Batch learning, also known as offline learning, is a machine learning technique where the model is trained on a fixed dataset that is known in advance. The model is trained on the entire dataset at once, as opposed to online learning, which trains the model on small batches of data over time.

#What is Online learning?

###Online learning, also known as incremental learning, is a machine learning technique where the model is trained on small batches of data over time. This is in contrast to batch learning, where the model is trained on a fixed dataset that is known in advance. In online learning, the model is updated continuously as new data becomes available.

###4. What is the learning rate
    
###The learning rate is a hyperparameter that controls the step size at each iteration while moving toward a minimum loss function during the training of a neural network. In other words, the learning rate determines how fast or slow a neural network model learns in the training process.
    
###5. out of core learning
    
###Out-of-core learning is a technique used in online learning where the dataset is too large to fit into memory. Instead, the data is read in small batches from the disk and processed incrementally. This allows for efficient training of models on large datasets without requiring expensive hardware.
